```python
# 刪除特殊符號
def 去特殊符號(t):
    import re
    # 去特殊符號
    t = re.sub(r"[\s+\.\!\/_,$%^*(+\"\']+|[+——！，。？、~@#￥%……&*（）]+", "",t)
    t = re.sub(r'[-，+、@,？⊂・ω{}◆●►❖★！!：\\/(=)…（）『』%《》$;；」:?=<>"／.&#"「_【】]', "",t)    
    # 去除數字
    t = re.sub(r"[0123456789]","",t)    
    # 去除英文  
    t = re.sub(r"[a-zA-Z]","",t) 
    t = re.sub(r" ","",t)
    t = re.sub(r"\n","",t)
    return t  
```
