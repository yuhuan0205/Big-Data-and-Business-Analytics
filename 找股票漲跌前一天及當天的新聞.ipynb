# 讀入資料

```python
import pandas as pd
import time,datetime
```

```python
news = pd.read_csv("國巨_news2.csv",encoding = 'big5')
stock = pd.read_excel("stock7%.xlsx")
```

```python
stock
```

# 把欄位轉換為datetime格式

```python
news['post_time'] = pd.to_datetime(news['post_time'])
news['post_time']
```

```python
stock['年月日'] = pd.to_datetime(stock['年月日'])
```

# 尋找股票漲跌一天之前的新聞

```python
data_title_up = []
data_content_up = []
data_date_up = []
data_title_down = []
data_content_down = []
data_date_down = []
for i in range(57):  #跑stock資料
    stock_date = stock["年月日"][i]
    for j in range(5026):  #跑news資料
        news_date = news['post_time'][j]
        if (news_date - stock_date ).days > 1 : #只找news在日期之前的
            break 
        if (stock_date - news_date).days == 0 : 
            if stock["漲跌幅"][i] >0:
                data_date_up.append(news["post_time"][j])
                data_title_up.append(news["title"][j])
                data_content_up.append(news["content"][j])
            else:
                data_date_down.append(news["post_time"][j])
                data_title_down.append(news["title"][j])
                data_content_down.append(news["content"][j])
        elif (stock_date - news_date).days == -1 and news_date.hour <= 13: #股票漲跌當天13.之前的新聞
            if stock["漲跌幅"][i] >0:
                data_date_up.append(news["post_time"][j])
                data_title_up.append(news["title"][j])
                data_content_up.append(news["content"][j])
            else:
                data_date_down.append(news["post_time"][j])
                data_title_down.append(news["title"][j])
                data_content_down.append(news["content"][j])
```

# 寫出資料

```python
DATA_up = pd.DataFrame({'date':data_date_up,'title':data_title_up,'content':data_content_up})
DATA_up.to_csv("國巨_news_up_7%.csv",encoding="big5")
DATA_down = pd.DataFrame({'date':data_date_down,'title':data_title_down,'content':data_content_down})
DATA_down.to_csv("國巨_news_down_7%.csv",encoding="big5")
```
